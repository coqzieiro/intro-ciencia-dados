```python
# biblioteca para visualização dos dados
import pandas as pd
import matplotlib.pyplot as plt

# link dos dados
url = "https://gist.github.com/aishwarya8615/89d9f36fc014dea62487f7347864d16a"

# ler a tabela a partir do link
df = pd.read_html(url)[0]

# exibe o dataframe
df.head(25)
```

```python
# seleciona as colunas 'Country' e 'Status' e remove duplicatas
paises = df[['Country', 'Status']].drop_duplicates()

# exibe o dataframe paises
paises
```

```python
# conta a frequência de cada status
status_counts = df['Status'].value_counts()
```

```python
# calcula a porcentagem dos países desenvolvidos
total_countries = status_counts.sum()
developed_percentage = (status_counts['Developed'] / total_countries) * 100
print(f"Porcentagem de países desenvolvidos: {developed_percentage:.2f}%")
```

```python
# cria o gráfico de setores
status_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribuição dos Países por Status')
plt.ylabel('')  # Remove o rótulo do eixo y
plt.show()
```

### **QUESTÃO 3**

```python
# link dos dados
url = "https://gist.github.com/aishwarya8615/89d9f36fc014dea62487f7347864d16a"

# ler a tabela a partir do link
df = pd.read_html(url)[0]
```

```python
# filtro dos dados dos países
countries = ['Ireland', 'Finland', 'Netherlands', 'Zambia', 'Angola']
df_filtered = df[df['Country'].isin(countries)]
```

```python
# plotagem da série temporal (expectativa de vida)
plt.figure(figsize=(10, 6))
for country in countries:
    country_data = df_filtered[df_filtered['Country'] == country]
    plt.plot(country_data['Year'], country_data['Life_expectancy'], label=country)

plt.xlabel('Ano')
plt.ylabel('Expectativa de Vida')
plt.title('Expectativa de Vida ao Longo do Tempo')
plt.legend()
plt.show()
```

```python
# países com menor e maior expectativa de vida em 2014
df_2014 = df[df['Year'] == 2014]
max_life_expectancy_country = df_2014.loc[df_2014['Life_expectancy'].idxmax()]['Country']
min_life_expectancy_country = df_2014.loc[df_2014['Life_expectancy'].idxmin()]['Country']

print(f"País com maior expectativa de vida em 2014: {max_life_expectancy_country}")
print(f"País com menor expectativa de vida em 2014: {min_life_expectancy_country}")
```

```python
# biblioteca para visualização dos dados
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests

# link dos dados
url = "https://www.worldometers.info/world-population/population-by-country"

# envia uma request
response = requests.get(url)

# faz o parsing do HTML
soup = BeautifulSoup(response.content, 'html.parser')

# encontra a tabela contendo os dados da população
table = soup.find('table', {'id': 'example2'})

# converte a tabela para um dataframe
df = pd.read_html(str(table))[0]

# exibe o df
df.head()
```

```python
# remover espaços extras dos nomes das colunas
df.columns = df.columns.str.strip()

# selecionar as colunas numéricas
df_numeric = df[['Population  (2024)', 'Yearly  Change', 'Net  Change',
                 'Density  (P/Km²)', 'Land Area  (Km²)', 'Migrants  (net)',
                 'Fert.  Rate', 'Med.  Age', 'Urban  Pop %']]

# substituir 'N.A.' por NaN
df_numeric = df_numeric.replace('N.A.', pd.NA)

# limpar e converter colunas que estão com strings em valores numéricos, ignorando NA
df_numeric['Yearly  Change'] = df_numeric['Yearly  Change'].str.replace('%', '', regex=False).astype(float)

# para a coluna 'Urban  Pop %', substituir 'N.A.' por NaN antes de converter para float
df_numeric['Urban  Pop %'] = df_numeric['Urban  Pop %'].str.replace('%', '', regex=False)

# converter a coluna 'Urban  Pop %' para float, lidando com valores ausentes
df_numeric['Urban  Pop %'] = pd.to_numeric(df_numeric['Urban  Pop %'], errors='coerce')

# agora, converter as demais colunas para float (caso necessário)
df_numeric = df_numeric.apply(pd.to_numeric, errors='coerce')

# calcular a matriz de correlação
correlation_matrix = df_numeric.corr()

# exibir a matriz de correlação
print("Matriz de Correlação:\n", correlation_matrix)

# identificar as variáveis menos correlacionadas (correlação próxima de 0)
least_correlated = correlation_matrix.unstack().sort_values().drop_duplicates()
print("Menores correlações:\n", least_correlated.head(10))
```

```python
# gerando o mapa de calor das correlações
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de Calor das Correlações')
plt.show()
```

```python
# transformar a matriz de correlação em um DataFrame e resetar o índice
correlacoes = correlation_matrix.unstack().reset_index()
correlacoes.columns = ['Variável 1', 'Variável 2', 'Correlação']

# remover as correlações de uma variável consigo mesma
correlacoes = correlacoes[correlacoes['Variável 1'] != correlacoes['Variável 2']]

# remover duplicatas, já que a matriz de correlação é simétrica
correlacoes['Pair'] = correlacoes.apply(lambda x: tuple(sorted([x['Variável 1'], x['Variável 2']])), axis=1)
correlacoes = correlacoes.drop_duplicates(subset='Pair').drop(columns='Pair')

# selecionar as 5 maiores correlações em valor absoluto
maiores_correlações = correlacoes.reindex(correlacoes['Correlação'].abs().nlargest(5).index)

maiores_correlações
```
