```python
# biblioteca para visualiza√ß√£o dos dados
import pandas as pd
import matplotlib.pyplot as plt

# link dos dados
url = "https://gist.github.com/aishwarya8615/89d9f36fc014dea62487f7347864d16a"

# ler a tabela a partir do link
df = pd.read_html(url)[0]

# exibe o dataframe
df.head(25)
```

```python
# seleciona as colunas 'Country' e 'Status' e remove duplicatas
paises = df[['Country', 'Status']].drop_duplicates()

# exibe o dataframe paises
paises
```

```python
# conta a frequ√™ncia de cada status
status_counts = df['Status'].value_counts()
```

```python
# calcula a porcentagem dos pa√≠ses desenvolvidos
total_countries = status_counts.sum()
developed_percentage = (status_counts['Developed'] / total_countries) * 100
print(f"Porcentagem de pa√≠ses desenvolvidos: {developed_percentage:.2f}%")
```

```python
# cria o gr√°fico de setores
status_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribui√ß√£o dos Pa√≠ses por Status')
plt.ylabel('')  # Remove o r√≥tulo do eixo y
plt.show()
```

### **QUEST√ÉO 2**

```python
# link dos dados
url = "https://gist.github.com/aishwarya8615/89d9f36fc014dea62487f7347864d16a"

# ler a tabela a partir do link
df = pd.read_html(url)[0]
```

```python
# extrair a vari√°vel Life_expectancy
life_expectancy = df['Life_expectancy']
```

```python
# calcular a m√©dia e a vari√¢ncia
mean_life_expectancy = life_expectancy.mean()
variance_life_expectancy = life_expectancy.var()

print(f"M√©dia da expectativa de vida: {mean_life_expectancy:.2f}")
print(f"Vari√¢ncia da expectativa de vida: {variance_life_expectancy:.2f}")
```

```python
# construir o histograma
fig = plt.figure(figsize=(6,4))
num_bins = 10
n, bins, patches = plt.hist(life_expectancy, num_bins, facecolor='blue', alpha=0.5, density=True, edgecolor='black', linewidth=1.2)
plt.xlabel("Expectativa de Vida", fontsize=15)
plt.ylabel("Frequ√™ncia", fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title("Histograma da Expectativa de Vida")
plt.show()
```

### **QUEST√ÉO 3**

```python
# link dos dados
url = "https://gist.github.com/aishwarya8615/89d9f36fc014dea62487f7347864d16a"

# ler a tabela a partir do link
df = pd.read_html(url)[0]
```

```python
# filtro dos dados dos pa√≠ses
countries = ['Ireland', 'Finland', 'Netherlands', 'Zambia', 'Angola']
df_filtered = df[df['Country'].isin(countries)]
```

```python
# plotagem da s√©rie temporal (expectativa de vida)
plt.figure(figsize=(10, 6))
for country in countries:
    country_data = df_filtered[df_filtered['Country'] == country]
    plt.plot(country_data['Year'], country_data['Life_expectancy'], label=country)

plt.xlabel('Ano')
plt.ylabel('Expectativa de Vida')
plt.title('Expectativa de Vida ao Longo do Tempo')
plt.legend()
plt.show()
```

```python
# pa√≠ses com menor e maior expectativa de vida em 2014
df_2014 = df[df['Year'] == 2014]
max_life_expectancy_country = df_2014.loc[df_2014['Life_expectancy'].idxmax()]['Country']
min_life_expectancy_country = df_2014.loc[df_2014['Life_expectancy'].idxmin()]['Country']

print(f"Pa√≠s com maior expectativa de vida em 2014: {max_life_expectancy_country}")
print(f"Pa√≠s com menor expectativa de vida em 2014: {min_life_expectancy_country}")
```

```python
# biblioteca para visualiza√ß√£o dos dados
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests

# link dos dados
url = "https://www.worldometers.info/world-population/population-by-country"

# envia uma request
response = requests.get(url)

# faz o parsing do HTML
soup = BeautifulSoup(response.content, 'html.parser')

# encontra a tabela contendo os dados da popula√ß√£o
table = soup.find('table', {'id': 'example2'})

# converte a tabela para um dataframe
df = pd.read_html(str(table))[0]

# exibe o df
df.head()
```

```python
# remover espa√ßos extras dos nomes das colunas
df.columns = df.columns.str.strip()

# selecionar as colunas num√©ricas
df_numeric = df[['Population  (2024)', 'Yearly  Change', 'Net  Change',
                 'Density  (P/Km¬≤)', 'Land Area  (Km¬≤)', 'Migrants  (net)',
                 'Fert.  Rate', 'Med.  Age', 'Urban  Pop %']]

# substituir 'N.A.' por NaN
df_numeric = df_numeric.replace('N.A.', pd.NA)

# limpar e converter colunas que est√£o com strings em valores num√©ricos, ignorando NA
df_numeric['Yearly  Change'] = df_numeric['Yearly  Change'].str.replace('%', '', regex=False).astype(float)

# para a coluna 'Urban  Pop %', substituir 'N.A.' por NaN antes de converter para float
df_numeric['Urban  Pop %'] = df_numeric['Urban  Pop %'].str.replace('%', '', regex=False)

# converter a coluna 'Urban  Pop %' para float, lidando com valores ausentes
df_numeric['Urban  Pop %'] = pd.to_numeric(df_numeric['Urban  Pop %'], errors='coerce')

# agora, converter as demais colunas para float (caso necess√°rio)
df_numeric = df_numeric.apply(pd.to_numeric, errors='coerce')

# calcular a matriz de correla√ß√£o
correlation_matrix = df_numeric.corr()

# exibir a matriz de correla√ß√£o
print("Matriz de Correla√ß√£o:\n", correlation_matrix)

# identificar as vari√°veis menos correlacionadas (correla√ß√£o pr√≥xima de 0)
least_correlated = correlation_matrix.unstack().sort_values().drop_duplicates()
print("Menores correla√ß√µes:\n", least_correlated.head(10))
```

```python
# gerando o mapa de calor das correla√ß√µes
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de Calor das Correla√ß√µes')
plt.show()
```
