### **QUESTÃO 2**

```python
# importa a biblioteca para manipulação de arrays
import numpy as np

# abrir arquivo CSV
data = pd.read_csv('iris-with-errors.csv', header=(0))

# exibe o tamanho do dataframe
print(data.shape)

# retorna as 25 primeiras linhas
data.head(25)
```

```python
# substitui os caractéres '?' por NaN
data = data.replace('?', np.nan)

# retorna as 25 primeiras linhas
data.head(25)
```

```python
# convertemos o dataframe para o formato Numpy e ignoramos a última coluna (classe)
X = np.array(data[data.columns[0:data.shape[1]-1]], dtype = float)

# função nanmean que calcula a média ignorando os NaN
averages = np.nanmean(X, axis = 0)

# armazena os valores no vetor
for i in np.arange(0, X.shape[0]):
    for j in np.arange(0, X.shape[1]):
        if(np.isnan(X[i,j]) == True): # verificar se é uma entrada com NaN
            X[i,j] = averages[j] # insere a média
print(X)
```

```python
# imprime as médias
formatted_averages = [f"{value:.2f}" for value in averages]
print(formatted_averages)
```

### **QUESTÃO 4**

```python
# abrir arquivo CSV
data = pd.read_csv('Advertising.csv', header=(0))
data.head(10)
```

```python
# obtem a matriz de correlação dos dados
corr = data.corr()

# Mostra a matriz
plt.figure(figsize=(7, 7))

# imshow é usado para mostrar imagens
plt.imshow(corr, cmap='Blues', interpolation='none', aspect='auto')

# mostra a barra lateral de cores
plt.colorbar()

# inclui o nome das variáveis
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns);
plt.suptitle('Correlation between variables', fontsize=15, fontweight='bold')
plt.grid(False)
plt.show()
```

```python
# exibe o valor da correlação dos dados
print(corr)
```

```python
# define uma correlação mínima
p = 0.75
var = []

# calcula os elementos mais correlacionados
for i in corr.columns:
    for j in corr.columns:
        if(i != j):
            if np.abs(corr[i][j]) > p: # se maior do que |p|
                var.append([i,j])
print('Variáveis mais correlacionadas:\n', var)
```
